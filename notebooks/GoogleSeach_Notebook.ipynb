## 2. Using the Google Search API

```
from googlesearch import search

# Docs: https://python-googlesearch.readthedocs.io/en/latest/
```

```
query = "penfolds 2017 max chardonnay"
```

```
def get_google_search_result(query, language="en", n_result=15):
  print("query: " + query)
  return [result for result in search(
      query,             # The query you want to run
      tld="com",         # The top level domain
      lang=language,     # The language e.g. english
      num=20,            # Number of results per page
      start=0,           # First result to retrieve
      stop=n_result,     # Last result to retrieve
      pause=2.0,         # Lapse between HTTP (web) requests
      )]
```

```
results_list = get_google_search_result(query)
```
